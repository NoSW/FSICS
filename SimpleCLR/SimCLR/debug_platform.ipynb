```python
import matplotlib.pyplot as plt
import numpy as np

data = {}
way_list = [1, 2, 5, 10, 20,50,100 ]
shot_list = [0, 1, 5, 10, 15, 20, 30, 50, 70, 100]
model_list = []

with open("0410") as f:
   for line in f:
        tokens = line.split(' ')
        model = tokens[0]
        ws = tokens[1]
        acc = tokens[2]
        if model not in model_list:
            model_list.append(model)
        if ws not in data:
            data[ws] = {}
        data[ws][model] = float(acc)


for way in way_list:
    data[f"{way}way-0shot"] = {}
    for model_name in model_list:
        data[f"{way}way-0shot"][model_name] = 1.0/way

for shot in shot_list:
    data[f"1way-{shot}shot"] = {}
    for model_name in model_list:
        data[f"1way-{shot}shot"][model_name] = 1.0

for way in way_list:
    for model in model_list:
        y = []
        for shot in shot_list:
            y.append(data[f"{way}way-{shot}shot"][model])
        y = np.array(y)
        plt.ylim(0.0, 1.0)
        plt.plot(np.array(shot_list), y, '^-',marker='<', label=f"{model}")

    plt.title(f'{way}-way,n-shot')
    plt.xlabel('shot')
    plt.ylabel('accuracy')
    plt.legend()
    plt.show()

# x = np.arange(0,10,0.2)
# y = np.random.rand(len(x))
# plt.plot(x[::4],y[::4], marker="o")
# plt.show()
```

```python
import math
def entropy(p):
    e = 0.0
    for p_i in p:
        e += p_i * math.log(p_i)
    return -e

print(entropy([0.000001,0.99999,0.00001]))
print(entropy([0.3333,0.3333,0.3333]))
```
